```python
import pandas as pd
import seaborn as sb
import matplotlib.pylab as plt


data_1 = pd.read_excel("customer_5000_rows_with_capital.xlsx")
data_2 = pd.read_csv("SuperMarket_Analysis_Corrected.csv")
```

```python
data_1["Sales"] = data_1["Quantity"] * data_1["UnitPrice"]
data_1.columns
```

```python
data_1
```

```python
data_1["InvoiceDate"] = pd.to_datetime(data_1["InvoiceDate"])
data_1['Date'] = data_1['InvoiceDate'].dt.strftime('%Y-%m-%d')
data_1["Time"] = data_1["InvoiceDate"].dt.time
data_2["Date"] = pd.to_datetime(data_2["Date"])
data_2['Date'] = data_2['Date'].dt.strftime('%Y-%m-%d')
data_2['Time'] = pd.to_datetime(data_2['Time']).dt.strftime('%H:%M:%S')
```

```python
data_cleaned_1 = data_1.drop('InvoiceDate',axis=1)
data_cleaned_1
```

```python
data_cleaned_2 = data_2
data_cleaned_2
```
